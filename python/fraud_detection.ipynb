# Feature Engineering

```python
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

df['is_outlier'] = ((df['Amount'] > upper_bound) | (df['Amount'] < lower_bound)).astype(int)
```

```python
df['risk_score'] = (
    (df['Amount'] > df['Amount'].quantile(0.99)).astype(int) +
    df['is_duplicate'] +
    (df['Time'] > df['Time'].quantile(0.95)).astype(int)
)
```

```python
df['is_outlier'].value_counts()
```

```python
df['risk_score'] += df['is_outlier']
```

```python
import pandas as pd

# Compare fraud rate within each group
outlier_fraud_check = df.groupby('is_outlier')['Class'].value_counts(normalize=True).unstack()
print(outlier_fraud_check)
```

```python
df['hour'] = (df['Time'] // 3600) % 24
```

```python
risk_fraud = df.groupby('risk_score')['Class'].mean()
print(risk_fraud)
```

```python
df['is_night'] = df['hour'].apply(lambda x: 1 if (x >= 0 and x <= 6) else 0)
```

```python
df['risk_score'] += (df['is_night'] == 1).astype(int)  # Increase risk for night txns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Amount_norm'] = scaler.fit_transform(df[['Amount']])
```

```python
def time_window(hour):
    if 0 <= hour < 6:
        return 'Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

df['Time_Window'] = df['hour'].apply(time_window)

fraud_by_window = df.groupby('Time_Window')['Class'].mean()
print(fraud_by_window)
```

```python
“Outliers in transaction amount show a significantly higher fraud rate (0.29%) compared to non-outliers (0.16%), nearly double. 
This supports including is_outlier as a key feature in risk scoring and prioritization.”
```

# Predictive Modeling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Time_scaled', 'Amount_scaled']] = scaler.fit_transform(df[['Time', 'Amount']])
```

```python
from sklearn.model_selection import train_test_split
X = df[['Amount_norm', 'risk_score', 'is_duplicate', 'is_outlier', 'hour', 'is_night']]
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)

```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', RandomForestClassifier(n_estimators=100))
])
pipeline.fit(X_train, y_train)
```

```python
importances = rf.feature_importances_
feature_names = X.columns
sns.barplot(x=importances, y=feature_names)
plt.title('Feature Importances')
plt.show()
```

```python
df['fraud_probability'] = model.predict_proba(X)[:, 1]
df['predicted_fraud'] = model.predict(X)
```

```python
df['final_risk'] = df['risk_score'] + (df['fraud_probability'] > 0.5).astype(int)
df['priority_review'] = (df['fraud_probability'] > 0.3) & (df['risk_score'] >= 2)
```
