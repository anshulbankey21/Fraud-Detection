```python
import numpy as np
Q1 = df["amount"].quantile(0.25)
Q3 = df["amount"].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df  = df[(df["amount"]  >= lower_bound) & (df["amount"]<=upper_bound)]
```

```python
df['Time_HHMM'] = df['Hour'].astype(str).str.zfill(2) + ':' + df['Minute'].astype(str).str.zfill(2)
```

```python
# Ensure 'Time' is numeric
df['time'] = pd.to_numeric(df['time'], errors='coerce')

# Extract hour and minute
df['Hour'] = (df['time'] // 3600).astype(int)
df['Minute'] = ((df['time'] % 3600) // 60).astype(int)
```

```python
df['hour'] = (df['time'] // 3600).astype(int)
```

```python
df.columns=df.columns.str.lower()
```

```python
def time_window(hour):
    if 0 <= hour < 6:
        return 'Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

df['Time_Window'] = df['hour'].apply(time_window)

fraud_by_window = df.groupby('Time_Window')['class'].mean()
print(fraud_by_window)
```

```python
# Convert seconds to hour (integer 0-23)
df['hour'] = (df['time'] // 3600).astype(int) % 24  # %24 to wrap around if > 24h

# Define function to categorize time of day
def get_time_of_day(hour):
    if 5 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 16:
        return 'Afternoon'
    elif 17 <= hour <= 20:
        return 'Evening'
    else:
        return 'Night'

# Apply function to create new column
df['Time_of_Day'] = df['hour'].apply(get_time_of_day)
```
